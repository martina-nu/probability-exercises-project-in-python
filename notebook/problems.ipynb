### Exercise 2

A box contains 10 white balls, 20 reds and 30 greens. Draw 5 balls with replacement. What is the probability that:

a. 3 white or 2 red

b. All 5 are the same color


We will use the same system here. We will pick our balls during each round and count the number of times that:

a. 3 white or 2 red 

b. All 5 are the same color.

```python
#import libraries
import numpy as np

d = {}

for i in range(60):
    if i < 10:
        d[i] = 'White'
    elif i > 9 and i < 30:
        d[i] = 'Red'
    else:
        d[i] = 'Green'  
  

#Initialize important variables

n_simulations = 1000
counter_a = 0 # 3 white and 2 red
counter_b = 0 # All 5 the same color


for i in range(n_simulations):
    
    #make a list of the colors that we choose
    balls = []

    for i in range(5): #Pick 5 random balls
        balls.append(d[np.random.randint(0,60)]) # desde un valor inferior incluido, hasta un valor superior sin incluir (son 60 bolas)
    
    #print(balls)
    
    #Convert it to a numpy array
   
    np_balls = np.array(balls)
    #print(np_balls)

    #find the number of each taht we picked

    white = 0
    red = 0
    green = 0

    for color in np_balls:
        if color == "White":
            white +=1
        if color == "Red":
            red +=1
        if color == "Green":
            green +=1

   #print(white,red,green)
    
    #Keep track if the combination met the criteria

    if white == 3 and red == 2:
        counter_a +=1
    if white == 5 or red == 5 or green == 5:
        counter_b +=1


    #print(counter_a,counter_b)

    
#print 'the probability of 3 white and 2 red is' in percentage

print('The probability of 3 white and 2 red is:', round(counter_a/n_simulations*100,2),'%')

#print 'the probability of all the same color is' in percentage   

print('The probability of all the same color is:', round(counter_b/n_simulations*100,2),'%')    
```
